## Chapter 4. (이어서) 중첩조건문

- nested counditional
- if 블록 안에 또다른 if 블록이 있는 경우
- 조건이 2개가 들어가는 것
- 들여쓰기로 논리를 구분한다
- 중첩조건문을 -> 간단하게 바꿀 수 있으면 최대한 바꾼다        

```python
# 변수 x가 y와 같으면 'same', y보다 작으면 'small', y보다 크면 'big'을 출력

# 안 좋은 예 1) 

x = 0
y = 1

if x == y:
    print("same")
else:
    if x < y:
        print("small")
    else:
        print("big")
```

```python
# 좋은 예 1) 단일 조건문으로 바꾸기

if x == y:
    print("same")
elif x < y:
    print("small")
else:
    print("big")
```

```python
# 변수 x의 값이 0과 같거나 작으면 'small'
# 0보다 크면서 10보다 작으면 'big', 10보다 크면 'too_big' 출력


# 안 좋은 예 2)
 
if x > 0:
    if x < 10:
        print("big")
    else:
        print("too_big")
else:
    print("small")
    
# 0 < x < 10
# 중첩조건문이 가능하지만 -> 바꿀 수 있으면 최대한 바꿔야 함
```

```python
# 좋은 예 2 - 1) 논리연산자 사용해 여러 조건을 한 줄에 넣기

if (x > 0) and (x < 10):
    print("big")
elif x <= 0:
    print("small")
else:
    print("too_big")
    
# 좋은 예 2 - 2) 파이썬에서만 가능
if 0 < x < 10:
    print("big")
elif x <= 0:
    print("small")
else:
    print("too_big")
```

## 논리연산자

- 비교연산자가 여러번 사용될 때 사용하는 것
- and, or, not
    - A and B (A & B): A도 참이고 B도 참일 때 True
    - A or B (A | b): A 혹은 B, 둘 중 하나만 참이면 True
    - not A: A가 False일 때 -> True, True -> False 반환
- 비교연산자의 괄호는 가독성을 위해 (optional)
- 파이썬에서만 가능한 표현: 0 < x < 10

## in ( 또는 not in)
- membership operator
- 어떤 요소가 그 안에 있는지 확인하고 싶다 (주의할 것)
- 비슷한 조건이 여러번 반복될 때 사용 가능

```python
letter = 'k'

# letter가 모음인지 확인하고 싶다 a, e, i, o, u
# if letter == 'a' or letter == 'e' ...

if letter in ['a', 'e', 'i', 'o', 'u']:  
    print("Yes")
else:
    print("No")
    
# 리스트 요소에 letter와 일치하는 것이 있는가
```

```python
# ***** 주의 *****
# word에 모음이 있는지 확인하고 싶을 때

# 틀린 예
word = 'abcde'

if word in ['a', 'e', 'i', 'o', 'u']:
    print("Yes")
else:
    print("No")

# word 안에 모음이 있는지 확인하는 코드가 아니다

# word 자체가 모음인가?를 확인하는 코드이다.
# word == a, e, i, o, u 라는 의미
```

```python
# 맞는 예

word = 'abcde'

if "a" or "e" or "i" or "o" or 'u' in word:
    print("Yes")
else:
    print("No")
```

## 바다코끼리 연산자
- walrus operator
- 할당과 테스트를 동시에 하는 기능
- := (콜론과 equal)

```python
tweet_limit = 200
tweet_string = 'blah' * 50 #공백없이 스트링 4 * 50개
tweet_string
```

```python
diff = tweet_limit - len(tweet_string)
if diff > 0:
    print("가능")
else:
    print("불가능")
```

```python
# 바다코끼리 연산자 사용해서 간단하게 표현

tweet_limit = 200
tweet_string = 'blah' * 50 

if diff := tweet_limit - len(tweet_string) > 0:
    print("가능")
```

## 5.4 Slicing
- 지정 범위의 `문자열`을 추출하는 것
- 문법: 대괄호 + 오프셋 이용 [offset]
    - [start:end:step]
    - 오프셋은 항상 0부터 시작
- start는 포함, end는 불포함 (크거나 같고, 작다) 
- start가 0일 경우 생략한다.
    - 예: ~4까지 출력하라 => [:5]
    - 처음부터 끝까지 2스텝씩 => {::2]
    - 처음부터 끝까지 역순으로 찍기 => [::-1]
- 오프셋은 거꾸로도 가능하다  
    0 1 2 3 4   
    h e l l o  
   -5      -1 (-1부터 시작한다)

```python
sent, sent[1] # 2번째 글자
```

```python
sent[3:4] #3만 나옴
```

```python
#처음부터 4까지
sent[:4], sent[0:4], sent[:5]
# 답은 [:5]
```

```python
sent2 = 'hello world'  #문자열은 공백도 offset으로 들어감
len(sent2) #length
```

```python
# 처음부터 끝까지 2스텝씩 출력하기
sent[::2]
```

```python
# 역순으로 출력
hello = 'hello'
hello[::-1]
```

```python
# 문자열은 불변이다
hello[2] = 'L' #(대문자로 바꾸고 싶어도 error)
```

```python
# 정렬(오름차순, reverse 내림차순)
sorted(cars)
sorted(cars, reverse = True)
```

```python
# Q1.

# 총 토큰 수 len(Thursday.tokens)
# 중복제외 토큰 수 len(set(Thursday.tokens))

# lexical diversity
len(set(Thursday.tokens)) / len(Thursday.tokens)
```

```python
# Q2 - 1)
sorted_Thursday = sorted(set(Thursday.tokens), reverse = True)[:10]
print(sorted_Thursday)

# Q2 -2)
for i in range(len(sorted_Thursday)):
    
    if 'z' in sorted_Thursday[i]:
        print(sorted_Thursday[i].upper())
        
    elif 'z' not in sorted_Thursday[i] and len(sorted_Thursday) >= 4:
        print(sorted_Thursday[i][:len(sorted_Thursday[i]) - 1] + sorted_Thursday[i].upper()[-1])
    
    else:
        print(sorted_Thursday)
```
