### 리스트 컴프리헨션 list comprehension
- 포함형/지능형/축의형 리스트
- 하나의 리스트에서 그걸 포함하여 다른 리스트를 만든다
- list + for loop
- [A for A' in C]
    - C에 있는 B에서 A를 가져와라 => 리스트에 담아서 반환
    - A: 내가 원하는 출력 형태
    - A': C의 형태
    - C: 반복가능한 자료형 (주로 리스트)
    
- 조건 추가: [A for A' in C if ...]

```python
print([color for color in colors])
print([color[0] for color in colors])
```

### NLTK
- frequency distribution : 단어가 몇 번 출현했는지 (빈도수)
    - 단어 개수
    - 단어 빈도수
- FreqDist(text)

```python
fdist = FreqDist(text1) # cased (즉, The의 개수와 the 개수가 다름)
fdist
```

```python
fdist.most_common(10) #상위 10개의 빈번한 것
```

```python
print(fdist['that']) # that이 몇 번 나왔는지 = text1.count('that')
print(fdist.freq('that')) # 몇퍼센(비율)트 차지하는지)
```

```python
fdist.plot(50, cumulative = True)

# 가장 빈번한 부분 -> least common
# 중복해서 쌓음 = 곡선이 완만할 수록 덜 중복

# pip install matplolib
```

```python
# 가장 빈번하지 않음 -> 1번만 나타나는 단어들
fdist.hapaxes()
len(fdist.hapaxes())
```

```python
# 알파벳인 것만 리스트에 담아보기, 중복없이
[w for w in fdist.hapaxes() if w.isalpha() == True ] # 주의 아님!
[w for w in fdist.hapaxes() if w.isalpha()] # 이렇게 
```

### 리스트 <-> 스트링
- split(): string.split(seperator를 지정)
    - default = ' ' (공백)
- join(): seperator.join(string)

```python
sent = 'I, have, a, car'

# 이렇게 변환 -> ['I','have', 'a', 'car'] #tokens
sent.split(',')
# 단어 사이에 공백 들어감
```

```python
# List 사이 공백 없애고 싶을 때
sent_tokens = [word.strip() for word in sent.split(',')]
sent_tokens
```

```python
# 다시 한 문장으로 붙이고 싶을 때 -> join
'-'.join(sent_tokens)
```
